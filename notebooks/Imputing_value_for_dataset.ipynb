#Imputing Absorption

```python
imp_abs = IterativeImputer(random_state=42)
imp_abs.fit(np.array(start_db['Absorption max (nm)']).reshape(-1,1))
change_nm_abs = imp_abs.transform(np.array(start_db['Absorption max (nm)']).reshape(-1,1))
change_nm_abs = change_nm_abs.flatten()
start_db['Absorption max (nm)'] = change_nm_abs
```

#Imputing Emission

```python
imp_em = IterativeImputer(random_state=42)
imp_em.fit(np.array(start_db['Emission max (nm)']).reshape(-1,1))
change_nm_em = imp_em.transform(np.array(start_db['Emission max (nm)']).reshape(-1,1))
change_nm_em = change_nm_em.flatten()
start_db['Emission max (nm)'] = change_nm_em
```

#Imputing Quantum yield

```python
imp_y = IterativeImputer(random_state=42)
imp_y.fit(np.array(start_db['Quantum yield']).reshape(-1,1))
change_y = imp_em.transform(np.array(start_db['Quantum yield']).reshape(-1,1))
change_y = change_y.flatten()
start_db['Quantum yield'] = change_y
```

#Imputing Lifetime

```python
imp_lt = IterativeImputer(random_state=42)
imp_lt.fit(np.array(start_db['Lifetime (ns)']).reshape(-1,1))
change_lt = imp_em.transform(np.array(start_db['Lifetime (ns)']).reshape(-1,1))
change_lt = change_lt.flatten()
start_db['Lifetime (ns)'] = change_lt
```
